EDA DATOS CATEGÓRICOS

```python
def graficos_eda_categoricos(cat):
    
    #Calculamos el número de filas que necesitamos
    from math import ceil
    filas = ceil(cat.shape[1] / 2)

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat 

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

```python
graficos_eda_categoricos(data.select_dtypes('O'))
```

EDA VARIABLES NUMÉRICAS

```python
def estadisticos_cont(num):
    #Calculamos describe
    estadisticos = num.describe().T
    #Añadimos la mediana
    estadisticos['median'] = num.median()
    #Reordenamos para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(estadisticos)
```

```python
estadisticos_cont(data.select_dtypes('number'))
```

Según el estudio "Cost of Turnover" del Center for American Progress:

* El coste de la fuga de los empleados que ganan menos de 30000 es del 16,1% de su salario

* El coste de la fuga de los empleados que ganan entre 30000-50000 es del 19,7% de su salario

* El coste de la fuga de los empleados que ganan entre 50000-75000 es del 20,4% de su salario

* El coste de la fuga de los empleados que ganan más de 75000 es del 21% de su salario

```python
#Creamos una nueva variable salario_ano empleado
data["salario_ano"] = data.salario_mes.transform(lambda x: x*12)
data[['salario_mes', 'salario_ano']]
```

```python
#Calculamos el impacto ecónomico de cada empleado que deja la empresa
condiciones = [(data['salario_ano'] <= 30000),
               (data['salario_ano'] > 30000) & (data['salario_ano'] <= 50000),
               (data['salario_ano'] > 50000) & (data['salario_ano'] <= 75000),
               (data['salario_ano'] > 75000)]

#Lista de resultados
resultados = [data.salario_ano * 0.161, data.salario_ano * 0.197, data.salario_ano * 0.204, data.salario_ano * 0.21]

#Aplicamos select
data['impacto_abandono'] = np.select(condiciones,resultados, default = -999)

data
```

```python
#Costo de los abandonos del último año
coste_total = data.loc[data.abandono == 1].impacto_abandono.sum()
coste_total
```

```python
# Porcentaje de representantes de ventas que se fueron el año pasado
total_repre_ventas = len(data[data["puesto"] == "Sales Representative"])
abandos_anno_pasado = len(data[(data.puesto == "Sales Representative") & (data.abandono == 1)])
porc_pasado = abandos_anno_pasado/total_repre_ventas
porc_pasado
```

```python
#Estimación de cuántos se irán este año
total_anno_actual = len(data[(data["puesto"] == "Sales Representative") & (data["abandono"] == 0)])
total_salen = int(porc_pasado * total_anno_actual)
total_salen
```

```python
#Si retenemos al 30%, cuánto nos ahorraríamos
ret = int(total_salen * 0.3)
ahorro = data.loc[(data["puesto"] == "Sales Representative") & (data["abandono"] == 0), 'impacto_abandono'].sum() * porc_pasado * 0.3

print(f"Podemos retener {ret} y ahorramos {ahorro}")
```

## MODELADO DE MACHINE LEARNING

```python
df_ml = data.copy()
```

```python
df_ml.info()
```

```python
#Convertir todas variables a numéricas
from sklearn.preprocessing import OneHotEncoder

#Variables categóricas
cat = df_ml.select_dtypes('O')

#Instanciación
ohe = OneHotEncoder(sparse_output=False)

#Entrenamos
ohe.fit(cat)

#Aplicamos
cat_ohe = ohe.transform(cat)

#Ponemos los nombres
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features = cat.columns)).reset_index(drop = True)
```

```python
cat_ohe.head()
```

```python
#Unión del data frama de variables categóricas con el data frame de variables numéricas
num = df_ml.select_dtypes('number').reset_index(drop=True)

df_ml = pd.concat([cat_ohe, num], axis = 1)
df_ml.head()
```

## DISEÑO DE LA MODELIZACIÓN

```python
#Separación de la variable a predecir y las variables predictoras
x = df_ml.drop(columns='abandono') #variables predictoras
y = df_ml["abandono"] #variable a predecir
```

```python
#Separación de los conjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size= 0.3)
```

```python
#Entrenando el modelo sobre train
from sklearn.tree import DecisionTreeClassifier

#Instanciar
ac = DecisionTreeClassifier(max_depth=4)

#Entrenar
ac.fit(train_x, train_y)
```

Predicción y validación sobre el conjunto de prueba

```python
pred = ac.predict_proba(test_x)[:,1]
pred[:20]
```

```python
#Evaluación
from sklearn.metrics import roc_auc_score
roc_auc_score(test_y,pred)
```

## Intrepretación

```python
from sklearn.tree import plot_tree

plt.figure(figsize = (50,50))

plot_tree(ac,
          feature_names= test_x.columns,
          impurity = False,
          node_ids = True,
          proportion = True,
          rounded = True,
          precision = 2);
```

```python
#Importancia de las variables
pd.Series(ac.feature_importances_,index = test_x.columns).sort_values(ascending = False).plot(kind = 'bar', figsize = (30,20));
```
