```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creación de un Data Frame
df = pd.DataFrame({"valores_1": np.random.rand(20)*100, "valores_2":np.random.rand(20)*100})
df
```

```python
#Creación del índice
indice = []
for i in range(len(df)):
    indice.append("alumno_" + str(i+1))

df.index = indice
df.head(4)
```

```python
#Creación de una nueva columna numerica
df["promedio"] = (df["valores_1"] + df["valores_2"])/2

df = df.round(2)

df.head()
```

```python
#Creación de una nueva columna categórica
df["estado"] = df["promedio"].apply(lambda x: "aprobado" if x >= 65 else "suspenso" if x >= 35 else "reprobado")
df.head()
```

```python
#Agregar una nota final
df['nota_final'] = df["estado"].apply(lambda x: round(np.random.randint(40,100),2) if 'suspenso' in x else "-")
df
```

```python
#Agregar promedio final
df["promedio_final"] = df.apply(lambda fila: fila["promedio"] if fila['estado'] in ['aprobado', 'reprobado'] 
                                else round((fila["promedio"] + fila["nota_final"])/2,2) if fila['estado'] == 'suspenso'
                                else 'reprobado', axis = 1)

df
```

```python
df["estado_final"] = df["promedio_final"].apply(lambda x: "APROBADO" if x >= 65 else "REPROBADO")
df
```

```python
estados_1 = df.groupby("estado").count()["promedio"]
estados_2 = df.groupby("estado_final").count()["promedio"]
print(estados_1, estados_2)
```

```python
fig, ax = plt.subplots()
ax.pie(x = estados_1.values, labels= estados_1.keys(), autopct='%1.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(x= estados_2.values, labels= estados_2.keys(), autopct='%1.1f%%')

plt.show()
```

```python
#Porcentaje de suspensos que aprueban
temp1 = df.groupby(["estado", "estado_final"]).count().reset_index()
total_suspensos = df.groupby("estado").count().loc["suspenso", 'promedio']
aprobados_suspensos = temp1.loc[2,'promedio']
promedio_apro_sus = round(aprobados_suspensos/total_suspensos*100,2)
promedio_apro_sus
```

```python
df["nota_final"] = pd.to_numeric(df["nota_final"], errors = "ignore")
df
```
