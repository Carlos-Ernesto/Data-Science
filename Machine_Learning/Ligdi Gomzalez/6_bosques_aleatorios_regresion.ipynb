```python
#Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
#Carga del dataset
boston = pd.read_csv('/BostonHousing.csv')
```

```python
#Separación de las variables independientes y objetivo
X_bar = boston[["RM"]]
y_bar = boston["MEDV"]
```

```python
#Gráfica de la relación de los datos
plt.scatter(X_bar, y_bar)

plt.show()
```

```python
#Separación de los conjuntos de entrenammiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_bar, y_bar, test_size=0.2)
```

```python
#Definición del modelo
modelo = RandomForestRegressor(n_estimators = 300, max_depth = 8)
```

```python
#Entrenamiento del modelo
modelo.fit(X_train, y_train)
```

```python
#Predicciones del modelo
y_pred = modelo.predict(X_test)
```

```python
#Comparativa de los datos predichos y de prueba
comparativa = pd.DataFrame({"y_test":y_test, "y_pred":y_pred})
comparativa
```

```python
X_grid = np.arange(min(X_test["RM"]), max(X_test["RM"]), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X_test, y_test)
plt.plot(X_grid, modelo.predict(X_grid), color = "red", linewidth = 2)

plt.show()
```

```python
#Precisión del modelo
modelo.score(X_train, y_train)
```
