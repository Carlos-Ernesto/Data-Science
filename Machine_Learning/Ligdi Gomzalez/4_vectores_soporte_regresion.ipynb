```python
#Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

```python
#Carga del dataset
boston = pd.read_csv("/BostonHousing.csv")
```

```python
#Separación de las variables dependientes y objetivo
X_svr = boston[['RM']]
y_svr = boston["MEDV"]
```

```python
#Gráfica de las variables
plt.scatter(X_svr, y_svr)

plt.show()
```

```python
#Separación de los datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_svr, y_svr, test_size = 0.2)
```

```python
#Definición del algoritmo
svr = SVR(kernel="linear", C=1.0, epsilon=0.2)
```

```python
#Entrenamiento del modelo
svr.fit(X_train, y_train)
```

```python
#Predicciones del modelo
y_pred =  svr.predict(X_test)
```

```python
#Comparación de valores
comparativa = pd.DataFrame({"y_pred":y_pred, "y_test":y_test})
comparativa
```

```python
#Gráfica de las valores predichos y de prueba
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color = "y", linewidth = 2)

plt.show()
```

```python
#Precisión del modelo
svr.score(X_train, y_train)
```
