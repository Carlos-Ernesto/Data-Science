```python
# Importación librerías

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Extracción del conjunto de datos

boston = pd.read_csv('BostonHousing.csv')
boston.info()
```

```python
# Separación de los conjuntos de entrenamiento y prueba

X = boston.iloc[:,5] # Para el ejemplo de regresión lineal simple solo se va a usar una columna de todo el dataset
y = np.array(boston["MEDV"]).reshape(-1,1)
print(type(y))
print(type(X))
```

```python
# Graficación de los datos

plt.scatter(X,y)
plt.xlabel('Número de habitaciones promedio')
plt.xlabel('Valor medio')

plt.show()
```

```python
# Separación de los conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2)
```

```python
# Difinición del algoritmo a utilizar
modelo = LinearRegression()
```

```python
# Entrenamiento del modelo
modelo.fit(X_train, y_train)
```
