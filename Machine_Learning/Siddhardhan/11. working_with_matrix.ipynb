```python
import numpy as np
```

Creating a matrix using a numpy

```python
matrix_1 = np.array([[2,3],[6,7]])
print(matrix_1)
print('\nMatrix shape:')
print(matrix_1.shape)
```

```python
matrix_2 = np.array([[10,35,43],[66,-43,-51],[26,6,-79]])
print(matrix_2)
print('\nMatrix shape:')
print(matrix_2.shape)
```

Creatring Matrices with Random Values

```python
random_matrix = np.random.rand(4,3)
print(random_matrix)
```

```python
# Creating a matrix with random integers
random_integer_matrix = np.random.randint(100,size=(4,5))
print(random_integer_matrix)
```

Matrix with all the values as 1

```python
matrix_3 = np.ones((2,3))
print(matrix_3)
```

```python
matrix_3 = np.ones((2,3), dtype=int)
print(matrix_3)
```

Null Matrix or Zero Matrix

```python
null_matrix = np.zeros((4,4))
print(null_matrix)
```

```python
null_matrix = np.zeros((7,3), dtype=int)
print(null_matrix)
```

Identity Matrix

```python
identity_matrix = np.eye(3,3)
print(identity_matrix)
```

```python
identity_matrix = np.eye(5,4, dtype=int)
print(identity_matrix)
```

Transpose Matrix

```python
# Matrix with random integers
a = np.random.randint(100, size = (4,5))
print(a)
```

```python
transpose_of_a = np.transpose(a)
print(transpose_of_a)
```
