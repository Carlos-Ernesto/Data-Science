```python
#Importing matplotlib library
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Generating random values
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

Plotting the data

```python
#Sine wave
plt.plot(x,y)
plt.show()
```

```python
#Cosine wave
plt.plot(x,z)
plt.show()
```

```python
#Adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')

plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)

plt.show()
```

```python
plt.plot(x,y, 'r+') #r= red += smbolo +

plt.show()
```

```python
plt.plot(x,y, 'g.') #puntos de color verde

plt.show()
```

```python
plt.plot(x,y, 'bx') #puntos de color azul

plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'y--')

plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages =['English', 'French', 'Spain', 'Chinese', 'German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')

plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages =['English', 'French', 'Spain', 'Chinese', 'German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels= languages, autopct = '%1.1f%%')

plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color = 'b')

plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection= '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)

ax.scatter(x,y,z, c =z , cmap = "Blues")

plt.show()
```
