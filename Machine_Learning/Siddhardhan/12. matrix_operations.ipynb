```python
import numpy as np
```

Matrix Adittion

```python
# Creating two matrices

a = np.array([[2,3],[4,5]])
b = np.array([[6,7],[8,9]])
```

```python
print(a)
print("\nMatrix Shape")
print(a.shape)
```

```python
print(b)
print("\nMatrix Shape")
print(b.shape)
```

Adding two matrices

```python
sum = a + b
print(sum)
```

```python
# Creating two matrices with random values
matrix_1 = np.random.randint(10, size = (3,3))
matrix_2 = np.random.randint(20, size = (3,3))

print('Matrix 1:\n', matrix_1,'\n\n','Matrix 2:\n', matrix_2)
```

```python
sum =  matrix_1 + matrix_1

print('Sum of matrices:\n', sum)
```

Matrix Subtaction

```python
a = np.array([[2,1],[4,3]])
b = np.array([[6,0],[7,9]])

difference = a - b
print(a,'\n\n',b, '\n')
print(difference)
```

Multiplying a Matrix by a Scaler

```python
x = 5
y = np.random.randint(10, size = (4,4))
print(y)
```

```python
product = np.multiply(x,y)
print(product)
```

Multiplying two Matrices

```python
# Creating two Matrices with random values
matrix_3 = np.random.randint(5, size = (3,3))
matrix_4 = np.random.randint(5, size = (3,4))
```

```python
print("Matrix 3:\n", matrix_3, "\n\nMatrix 4:\n", matrix_4)
```

```python
product = np.dot(matrix_3, matrix_4)
print(product)
print('\n')
print(product.shape)
```

Element wise Matrix Multiplication

```python
# Creating two Matrices with random values
matrix_5 = np.random.randint(1,5, size = (3,3))
matrix_6 = np.random.randint(1,5, size = (3,3))
```

```python
print("Matrix 5:\n", matrix_5, "\n\nMatrix 6:\n", matrix_6)
```

```python
product = np.multiply(matrix_5, matrix_6)
print(product)
```
