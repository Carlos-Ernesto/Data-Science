```python
#Importación de librerías
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#Creación de la semilla
np.random.seed(1)

#Creación 2 conjuntos con 50 variables de cada uno
x = np.linspace(0,50,50)
y = np.linspace(0,50,50)

x += np.random.uniform(-4,4,50)
y += np.random.uniform(-4,4,50)
```

```python
#Gráfica de los valoes
plt.scatter(x,y, color = "red")

plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
plt.grid(alpha = 0.3, color = "blue")

plt.show()
```

```python
#Convirtiendo el arreglo x de un arreglo plano a un arreglo de 1 dimensión
x = x.reshape((-1,1))
```

```python
#Modelo de regresión lineal
model = LinearRegression()

#entrenamiento del modelo
model.fit(x,y)
```

```python
#Error R^2
model.score(x,y)
```

```python
#Predicciones
y_pred = model.predict(x)
y_pred
```

```python
#Evalución de la métrica RMSE
RMSE = np.sqrt(mean_squared_error(y_pred,y))
RMSE
```

```python
#Gráfica de la línea de regresión lienal
plt.plot(x,y, "o", label = "Valores Reales")
plt.plot(x,y_pred, label = "Regresión Lineal")

plt.xlabel("Valores en X")
plt.ylabel("Valores en Y")
plt.legend()
plt.title("Regresión Lineal vs Valores Reales")

plt.grid(alpha = 0.3, color = "green")

plt.show()
```

```python
ads = pd.read_csv('Advertising.csv')
```

```python
#Visualización de la relación de los valores
sns.pairplot(ads, x_vars= ['TV', 'Radio', 'Newspaper'], y_vars= ["Sales"], size = 5, aspect=1)
plt.show()
```

```python
#Matriz de correlación
m = ads.loc[:, 'TV':].corr()
plt.figure(figsize=(10,10))
sns.heatmap(m, annot= True, cmap="Reds")
```

```python
#Separación de los conjunos de entrenamiento y práctica
x_train, x_test, y_train, y_test = train_test_split(ads.loc[:, ["TV", "Radio"]], ads["Sales"], test_size= 0.2, random_state=42)
```

```python
#Iniciar el modelo
model_2 = LinearRegression()

#Entrenamiento del modelo
model_2.fit(x_train, y_train)
```

```python
#R cuadrado
model_2.score(x_test, y_test)
```

```python
#Predicciones
y_pred = model_2.predict(x_test)
```

```python
#RMSE test 
RMSE = np.sqrt(mean_squared_error(y_pred,y_test))
RMSE
```
