### EXERCISE:

Predict Canada's per capita income in year 2020.

```python
#Importación de librerías
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Importación de los datos
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
df.describe()
```

```python
#Gráfica de los valores
plt.scatter(df.iloc[:,0], df.iloc[:,1])

plt.xlabel('Year')
plt.ylabel("Ingreso por capita")

plt.show()
```

```python
#Creación del modelo de regresión lineal
reg = linear_model.LinearRegression()

#Entrenamiento del modelo
reg.fit(df[['year']], df.iloc[:,1])
```

```python
reg.predict([[2020]])
```

```python
#Gráfica comparativa entre valores reales y predichos
%matplotlib inline

plt.scatter(df.iloc[:,0], df.iloc[:,1])
plt.plot(df.iloc[:,0], reg.predict(pd.DataFrame(df.iloc[:,0])), color = "red")

plt.xlabel('Year')
plt.ylabel("Ingreso por capita")

plt.show()
```
