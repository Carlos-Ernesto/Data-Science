```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier #Red neuronal
from sklearn.metrics import(accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, classification_report) 
```

```python
#Extrar los datos y la variable targetr
x, y = make_moons(n_samples=200) #Genera una muestra de 200 valores
y[:5]
```

```python
plt.scatter(x[:,0], x[:,1], c = y)

plt.grid()
plt.show()
```

```python
#Dividir los datos en train y test
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
#Modelo de la red neuronal
clf = MLPClassifier()
```

```python
#Entrenamiento de la red nueronal
clf.fit(x_train, y_train)
```

```python
#Accuracy de la red neuronal
clf.score(x_test, y_test)
```

```python
#Resultados predichos por la red neuronal
y_pred = clf.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
matriz_confusion = confusion_matrix(y_test, y_pred)
matriz_confusion
```

```python
#Cálculo de la precisión
precision = precision_score(y_test, y_pred)
print(precision, 21/26)
```

```python
#Cálculo del recall
recall = recall_score(y_test, y_pred)
print(recall, 21/22)
```

```python
f1 = f1_score(y_test, y_pred)
f1
```

```python
report = classification_report(y_test, y_pred)
print(report)
```
